# Index Replication

We are given $N$ stocks and an index.

Strategy: At rebalancing time $t$, we use return information over the period from $\left[t-obs, t-1\right]$ to compute portfolio allocations $w_t$. We hold this portfolio over the period $\left[t,t+hold\right]$. This is a simplified set-up to compare different loss measures and also to illustrate randomised splitting of training and test set without overlap.

Features are $N \times obs$ matrices. We treat them as single channel images, so we reshape them to $N \times obs \times 1 \times B$, where B is the batch size. We take $B=1$ for this exercise.

Let $r_s^p = \sum_jw_t^jr_{s}^j, s = t+1, t+2, \ldots, t+hold$ be portfolio returns and $r_s^I$ be the corresponding index returns. We explore several "closeness" measures:

1. mean squared error (MSE)
$$\frac{1}{hold}\sum_s \vert r_s^p - r_s^I \vert^2$$

2. maximum squared error (MXSE)
$$\max_s \vert r_s^p - r_s^I \vert^2$$

3. mean cumulative squared error (MCSE)
$$\frac{1}{hold}\sum_s\left(\frac{\sum_{u=1}^sr_{u}^p}{s} - \frac{\sum_{u=1}^sr_{u}^I}{s}\right)^2$$

4. maximum absolute deviation (MAD) 
$$\max_s \vert r_s^p - r_s^I \vert$$

This constant % allocation within a time slice simplifies transaction cost computation, though the between time slices rebalancing costs are ignored. Transaction cost for stock $i$ at time $t$ is

$$c_i \sum_i w^i \vert r_t^i - r_t^p \vert$$

```julia
using Flux, CSV, DataFrames, LinearAlgebra, Random, BSON, Statistics
using TensorBoardLogger, Logging, ProgressMeter
using Flux.Data: DataLoader
using MLUtils
```

## Helper Functions

### Features

```julia
function get_data(stx, idx, args)
    T = size(stx,2)
    rng = args.obs+1:args.hold:T-args.hold
    xs = [stx[:,t-args.obs:t+args.hold] for t ∈ rng]
    ys = [idx[t+1:t+args.hold] for t ∈ rng]
    n = length(xs)
    ids = collect(1:n)
    shuffle!(ids)
    ntrain = floor(Int, args.trainfrac * n)
    trg_ids, tst_ids = ids[1:ntrain], ids[ntrain+1:end]
    train_loader = zip(xs[trg_ids],ys[trg_ids])
    test_loader = zip(xs[tst_ids],ys[tst_ids])
    return train_loader, test_loader
end
```

### Loss Functions

```julia
mse(x) = sqrt(sum(abs2, x)/length(x))
mxse(x) = sqrt(maximum(x.^2))
mad(x) = maximum(abs.(x))
mcse(x,y) = sqrt(sum(abs2, cumsum(x) ./ (1:length(x)) .- cumsum(y) ./ (1:length(y)))/length(x))
```

```julia
function loss(m,x,y,args)
    f = reshape(x[:,1:args.obs],size(x,1),args.obs,1,1)
    w = m(f) |> transpose
    r = x[:,args.obs+2:end]
    tmp = w*r
    rp = tmp .- args.tcost * w * (r .- tmp)
    func = args.lossfn
    if func != mcse
        return func(vec(rp) .- y)
    end
    return func(vec(rp), y)
end
```

### Logging

```julia
function report(tblogger, model, data, epoch, args)
    trg_loss = mean([loss(model,x,y,args) for (x,y) in data[1]])
    tst_loss = mean([loss(model,x,y,args) for (x,y) in data[2]])   
    set_step!(tblogger, epoch)
    with_logger(tblogger) do
        @info "train" loss=trg_loss
        @info "test"  loss=tst_loss
    end
end
```

```julia
make_logid(args) = String(Symbol(args.lossfn))*
                "_obs_$(args.obs)_hold_$(args.hold)"*
                "_lr_$(args.η)_reg_$(args.λ)"*
                "_k_$(args.kernel_size)"*
                "_c_$(join(string.(args.channels),'_'))"
```

### CNN Builder

Create convolution block with convolution, normalisation, dropout and pooling layers, in that order.

```julia
function conv_block(cin,cout,k; p=0.5, act=relu)
    l = []
    kernel = (k,k)
    push!(l, Conv(kernel, cin => cout, act, pad=SamePad()))
    push!(l, BatchNorm(cout))
    push!(l, Dropout(p))
    push!(l, MaxPool(kernel))
    return l
end
```

String together convolution blocks, followed by a dense layer to produce allocation that are non-negative and sum to one.

```julia
function make_model(h, args)
    w, k, c = args.obs, args.kernel_size, args.channels
    p, act = args.dropout, args.actfn 
    layers = []
    kernel = (k,k)
    conv = conv_block(1,c[1],k,p=p, act=act)
    push!(layers, conv...)
    for (i, o) in zip(c, c[2:end])
        conv = conv_block(i,o,k,p=p, act=act)
        push!(layers, conv...)
    end
    conv_size = Flux.outputsize(layers, (w, h, 1); padbatch=true)
    linear = Dense(prod(conv_size), h, act)
    return Chain(layers..., Flux.flatten, linear, NNlib.softmax)
end
```

```julia
function load_model(check_point; kws...)
    args = Args(; kws...)
    logid = make_logid(args)
    savepath = joinpath(args.savepath, logid)
    modelpath = joinpath(savepath,"model_epoch_$(check_point).bson")
    BSON.@load modelpath model
    return model
end
```

### Training

```julia
function train(stx, idx; kws...)
    args = Args(; kws...)
    data = get_data(stx, idx, args)
    
    logid = make_logid(args)
    savepath = joinpath(args.savepath, logid)
    tblogger = TBLogger(savepath, tb_overwrite)
    
    model = make_model(size(stx,1), args)
    opt = Flux.setup(OptimiserChain(WeightDecay(args.λ), ADAM(args.η)), model)
    
    @showprogress for e in 1:args.epochs
        for (x,y) in data[1]
            gs = gradient(m->loss(m,x,y,args),model)
            Flux.update!(opt,model,gs[1])
        end
        if e % args.info_freq == 0
            report(tblogger, model, data, e, args)
        end
        if e % args.check_freq == 0
            modelpath = joinpath(savepath,"model_epoch_$e.bson")
            BSON.@save modelpath model
        end
    end
end
```

### Performance

```julia
function get_nav(stx, idx, model; kws...)
    args = Args(; kws...)
    T = size(stx,2)
    rng = args.obs+1:args.hold:T-args.hold
    xs = [stx[:,t-args.obs:t+args.hold] for t ∈ rng]
    ys = [idx[t+1:t+args.hold] for t ∈ rng]
    rep, tgt = [], []
    for (x,y) in zip(xs, ys)
        f = reshape(x[:,1:args.obs],size(x,1),args.obs,1,1)
        w = model(f) |> transpose
        r = x[:,args.obs+2:end]
        tmp = w*r
        rp = tmp .- args.tcost * w * (r .- tmp)
        append!(rep, rp)
        append!(tgt,y)
    end
    return (portfolio=rep,index=tgt,timeid=rng)
end
```

## Training Environment

Structure to hold default values for all the variables necessary for training.

```julia
Base.@kwdef mutable struct Args
    epochs::Int = 500
    η::Float32 = 1e-3
    λ::Float32 = 1e-5
    info_freq::Int = 10
    check_freq::Int = 20
    obs::Int = 30
    hold::Int = 30
    trainfrac::Float32 = 0.7
    tcost::Float32 = 0.005
    lossfn::Function = mse
    savepath::String = "./log"
    kernel_size::Int = 2
    channels::Vector{Int} = [4,8]
    actfn::Function = relu
    dropout::Float32 = 0.5
end 
```

## Experiment

Load stocks and index returns data. Returns are log returns.

```julia
df = CSV.read("hsi_rets.csv",DataFrame)
dropmissing!(df)
tid = df.Date
idx = df[:,2] |> Vector{Float32}
stx = df[:,3:end] |> Matrix{Float32} |> transpose
size(idx), size(stx)
```

Run experiments.

```julia
for fn in [mcse, mse, mad, mxse]
    train(stx, idx, lossfn=fn)
end
```

Compare index and tracking portfolio.

```julia
model = load_model(500, lossfn = mxse)
```

```julia
res = get_nav(stx,idx,model);
```

```julia
plot(cumsum(res.index), label="index")
plot!(cumsum(res.portfolio), label="replica")
```
